# Example 1-1
Code for trainin a basic Linear regression model

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Downloading data author's web
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X= lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
# Visualizing data with scatterplot
lifesat.plot(kind='scatter', grid=True, x="GDP per capita (USD)", y = "Life satisfaction",  )
```

```python
# Select a model
model = LinearRegression()
```

```python
# Model training
model.fit(X,y)
```

```python
# Make a new prediction
X_new = model.predict([[37_655.2]])
```

```python
X_new
```

```python
# Extra code: graphic the line
a0 = model.intercept_[0]
a1 = model.coef_[0][0]

x_values = np.linspace(25000, 60000, 50)
y_values = a0 + a1*x_values
```

```python
lifesat.plot(kind='scatter', grid=True, x="GDP per capita (USD)", y = "Life satisfaction",  )
plt.plot(x_values,y_values, color='green')
plt.show()
```
